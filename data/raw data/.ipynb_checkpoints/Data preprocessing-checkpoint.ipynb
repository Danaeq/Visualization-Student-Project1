```python
import numpy as np
import pandas as pd
```

```python
military_gdp_raw=pd.read_excel('Military expenditure (% of GDP).xls',header = 3,index_col = 0)
health_gdp_raw=pd.read_excel('Current health expenditure (% of GDP).xls',header = 3,index_col=0)
education_gdp_raw=pd.read_excel('Government expenditure on education, total (% of GDP).xls',header = 3,index_col=0)
gdp_raw=pd.read_excel('GDP (current US$).xls',header=3,index_col = 0)
population_raw=pd.read_excel('population.xls',header=3,index_col = 0)
```

```python
military_gdp = military_gdp_raw.ix[['United States','China','Russian Federation','United Kingdom','Germany','France', 'Italy','Mexico', 'Argentina','South Africa', 'Australia','Indonesia'],['2012','2013','2014','2015','2016']]
health_gdp = health_gdp_raw.ix[['United States','China','Russian Federation','United Kingdom','Germany','France', 'Italy','Mexico', 'Argentina','South Africa', 'Australia','Indonesia'],['2012','2013','2014','2015','2016']]
education_gdp = education_gdp_raw.ix[['United States','China','Russian Federation','United Kingdom','Germany','France', 'Italy','Mexico', 'Argentina','South Africa', 'Australia','Indonesia'],['2012','2013','2014','2015','2016']].fillna(0)
gdp = gdp_raw.ix[['United States','China','Russian Federation','United Kingdom','Germany','France', 'Italy','Mexico', 'Argentina','South Africa', 'Australia','Indonesia'],['2012','2013','2014','2015','2016']]
population = population_raw.ix[['United States','China','Russian Federation','United Kingdom','Germany','France', 'Italy','Mexico', 'Argentina','South Africa', 'Australia','Indonesia'],['2012','2013','2014','2015','2016']]
```

```python
military = military_gdp*gdp/100
health = health_gdp*gdp/100
education = education_gdp*gdp/100
```

```python
#military
```

```python
#health
```

```python
#education
```

```python
mean_military_raw = np.mean(military, axis=1)
mean_health_raw = np.mean(health,axis=1)
mean_education_raw = np.mean(education,axis=1)
mean_gdp_raw = np.mean(gdp,axis=1)

mean_military= pd.DataFrame(mean_military_raw)
mean_health= pd.DataFrame(mean_health_raw)
mean_education= pd.DataFrame(mean_education_raw)
mean_gdp= pd.DataFrame(mean_gdp_raw)

mean_military.columns = ['military']
mean_health.columns = ['health']
mean_education.columns = ['eductaion']
mean_gdp.columns = ['gdp']
```

```python
#mean_gdp
```

```python
summary_mean = [mean_military,mean_health,mean_education,mean_gdp]
mean_spending = pd.concat(summary_mean,axis=1)
mean_spending.head()
```

```python
military_pp = military/population
health_pp = health/population
education_pp = education/population
gdp_pp = gdp/population
military_pp
```

```python
military_2012= military.ix[:,['2012']]
military_2012.columns = ['military2012']
health_2012 = health.ix[:,['2012']]
health_2012.columns = ['health2012']
education_2012 = education.ix[:,['2012']]
education_2012.columns = ['education2012']

summary_2012 = [military_2012,health_2012,education_2012]
spending_2012 = pd.concat(summary_2012,axis=1)
#spending_2012
```

```python
military_2016= military.ix[:,['2016']]
military_2016.columns = ['military2016']
health_2016 = health.ix[:,['2016']]
health_2016.columns = ['health2016']
education_2016 = education.ix[:,['2016']]
education_2016.columns = ['education2016']

summary_2016 = [military_2016,health_2016,education_2016]
spending_2016 = pd.concat(summary_2016,axis=1)
#spending_2016
```

```python
with pd.ExcelWriter('cleaned data.xlsx') as writer: 
    military.to_excel(writer, sheet_name='military spending')
    health.to_excel(writer, sheet_name='health spending')
    education.to_excel(writer, sheet_name='education spending')
    mean_spending.to_excel(writer, sheet_name='mean_spending')
    gdp.to_excel(writer, sheet_name='gdp')
    military_pp.to_excel(writer, sheet_name='military spending per person')
    health_pp.to_excel(writer, sheet_name='health spending per person')
    education_pp.to_excel(writer, sheet_name='education spending per person')
    gdp_pp.to_excel(writer, sheet_name='gdp per person')
```
